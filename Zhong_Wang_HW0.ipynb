###### (f)	Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers.  (Here, use the iterrows() method to iterate each row of the dataframe and copy it to a dictionary. However, there is an easier way. Learn how the to_dict() method works by yourself later)

```python
data_dict = {'yearID': [], 'teamID': [], 'lgID': [], 'playerID': [], 'salary': []}

for label, row in data.iterrows():
    data_dict['yearID'].append(row.yearID)
    data_dict['teamID'].append(row.teamID)
    data_dict['lgID'].append(row.lgID)
    data_dict['playerID'].append(label)
    data_dict['salary'].append(row.salary)
```

```python
print(data_dict)
```

###### (g)	Create a dataframe using pd.DataFrameRead (Documentation Link) and from the dictionary created in (e). Then, change the header to "a", "b", "c", ... .

```python
# create a dataframe using the dict
df = pd.DataFrame(data=data_dict, columns={"yearID":"a", "teamID":"b", "lgID":"c", "playerID":"d", "salary":"e"})

# change the header to "a", "b", "c", ...
_ = df.columns=["a", "b", "c", "d", "e"]
```

### 5. Numpy

###### (a)	Create a 2-dimensional Python list object,  then convert it to a Numpy array object.

```python
list2d = [[0, 1, 0], [1, 2, 1]]

list2d = np.array(list2d)
print(list2d)
```

###### (b)	Examine the ndim, shape, size, dtype, itemsize, and data attributes of the numpy array object. Make sure you understand their functions.

```python
# ndim
dims = list2d.ndim
print(f"list2d has {dims} dimensions.")

# shape
rows, cols = list2d.shape
print(f"list2d has {rows} rows and {cols} cols.")

# size
num_items = list2d.size
print(f"list2d has {num_items} elements.")

# dtype
data_type = list2d.dtype
print(f"list2d is of type {data_type.name}.")

# itemsize
item_size = list2d.itemsize
print(f"An element in list2d is {item_size} bytes in size.")

# type
print(f"list2d is of type {type(list2d)}.")
```

###### (c)	Learn the dimension concept of an ndarray object by using reshape() and flatten() methods.

```python
# reshape
list2d_3by2 = list2d.reshape(3, 2)

print(f"list2d reshaped into a 3 by 2 dimensions array:\n {list2d_3by2}\n")

# flatten
list2d_1 = list2d.flatten()
print(f"list2d flattened into a 1 dimensional array:\n {list2d_1}\n")
```

###### (d)	Understand how the slice operation works for 1-D arrays and 2-D arrays and practice by yourself.

```python
# slicing 1D array
list2d_1_last3 = list2d_1[3:]

print(f"list2d_1 sliced to last 3 elements:\n {list2d_1_last3}\n")

# slicing 2D arrays
array2d = np.arange(16).reshape(4, 4)

print(f"Array before slicing: \n {array2d}\n")

array2d_middle = array2d[1:3:1, 1:3:1]
print(f"Array after slicing: \n {array2d_middle}\n")
```

###### (e)	Learn operations on ndarray by examining the argmin(), argmax(), min(), max(), mean(), sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), and mod() methods. Make yourself comfortable with these methods.

```python
##### create an array from list
a = np.array([[10, 14, 12], [15, 11, 13]])
print("a\n", a, end="\n\n")

b = np.array([[2, 4, 6], [8, 10, 12]])
print("b\n", b, end="\n\n")

c = np.arange(6)
print("c\n", c)
```

```python
# argmin, min
print(f"index of minimum element {np.min(a)}  in the array: {a.argmin()}")
print(f"index of minimum element {min(a.flatten())} in the array (x-axis): {np.argmin(a, 0)}")

# argmax, max
print(f"index of maximum element {np.max(a)} in the array: {a.argmax()}")
print(f"index of maximum element {max(a.flatten())} in the array (y-axis): {np.argmax(a, 1)}")

# mean
print(f"Mean of the array: {np.mean(a)}")
print(f"Mean of the array (x-axis): {np.mean(a, 0)}")
print(f"Mean of the array (y-axis): {np.mean(a, 1)}")

# sum
print(f"sum of the array: {np.sum(a)}")
print(f"sum of the array (x-axis): {np.sum(a, 0)}")
print(f"sum of the array (y-axis): {np.sum(a, 1)}")

# std
print(f"Standard deviation from mean of the array: {np.std(a)}")
print(f"Standard deviation from mean of the of the array (x-axis): {np.std(a, 0)}")
print(f"Standard deviation from mean of the of the array (y-axis): {np.std(a, 1)}")

# dot
print(f"a . b = {np.dot(a.ravel(), b.ravel())}")
print(f"b . a = {np.dot(b.ravel(), a.ravel())}")

# square
b_squared = np.square(b)
print(f"Square of array b (b^2):\n {b_squared}\n")

# sqrt
print(f"Square of array b (b^2):\n {b_squared}\n")

# abs
print(f"Square root of array b^2: \n {np.sqrt(b_squared)}\n")

# exp
print(f"Exponential of array c:\n {np.exp(c)}\n")

# sign
print(f"Signs of elements in c: {np.sign(c)}\n")

# mod
print(f"Floor division remainder of c by 5:\n {np.mod(c, 5)}\n")
```

###### (f)	Examine the arange(), ones(), zeros(), eye(), linspace(), and concatenate() methods. Make yourself comfortable with these methods.

```python
# arange
print(f"Array from a range of numbers:\n {np.arange(2, 14, 2).reshape(2, 3)}\n")

# ones
print(f"Array of ones:\n {np.ones((2, 3))}\n")

# zeros
print(f"Array of zeros:\n {np.zeros((2, 3))}\n")

# eye
print(f"Eyed array:\n {np.eye(3)}\n")

# linspace
print(f"Linear evenly spaced array:\n {np.linspace(0, 10, num=6).reshape(2, 3)}\n")

# concatenate
print(f"Default concatenation: a with b:\n {np.concatenate((a, b))}\n")
print(f"y-axis concatenation: a with b:\n {np.concatenate((a, b), 0)}\n")
print(f"x-axis concatenation: a with b:\n {np.concatenate((a, b), 1)}\n")
```

###### (a)	Create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid.

###### (b)	Create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot.

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
```

```python
fig = plt.figure()
ax = plt.subplot(111)
plt.title('x vs y1, y2, y3')
_ = ax.plot(x, y1, label='x against y1')
_ = ax.plot(x, y2, label='x against y2')
_ = ax.plot(x, y3, label='x against y3')

_ = ax.legend()
```

###### (c)	Create multiple subplots, play around with the figure size, text font/size.

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('Multiple subplots', fontsize=14)
fig.set_figwidth(16)

_ = ax1.plot(x, y1)
ax1.set_xlabel("x")
ax1.set_ylabel("y1")

_ = ax2.plot(x, y2)
ax2.set_xlabel("x")
ax2.set_ylabel("y2")

_ = ax3.plot(x, y3)
ax3.set_xlabel("x")
ax3.set_ylabel("y3")
```
